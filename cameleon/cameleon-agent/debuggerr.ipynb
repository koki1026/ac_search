```python
import gymnasium as gym
import cameleon_gym_env_path
import time

env = gym.make('CameleonPath-v0')

#環境を初期化
observation , _= env.reset(render_mode = 'human')

#アクションを指定回数実行
for _ in range(200):
    action = env.action_space.sample() #ランダムなアクションを取得
    print("action:; ", action)
    observation, reward, done, _, info = env.step(action) #アクションを実行
    if done:
        print("Episode finished after {} timesteps".format(_+1))
        observation , _ = env.reset(render_mode = 'human') #環境を初期化
    #1秒待機
    time.sleep(1)
```
